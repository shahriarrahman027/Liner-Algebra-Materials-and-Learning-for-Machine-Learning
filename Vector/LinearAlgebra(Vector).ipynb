```python
import numpy as np 
A=np.array([1,2,3,4,5])
distance=np.linalg.norm(A)
print(distance)
```

# Proving the Commutative Rule 

```python
import numpy as np
A=np.array([1,2,3])
B=np.array([4,5,6])
C=np.array([7,8,9])
AB=np.dot(A,B)
BA=np.dot(B,A)
if AB==BA:
    print("Commutative law is proven")
```

# Proving the Distributive law:

```python
A1=np.dot(A,B+C)
A2=np.dot(A,B)+np.dot(A,C)
if A1==A2:
    print("Distributive law is also proven")
else:
    print("NO,this formula is not priven with this example")
```
